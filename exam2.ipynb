```python
import numpy as np
X = np.matrix('1,.9,.8; 1,.5,.6; 1,.7,.7; 1,.6,.5')
Y = np.matrix('.95; .5; .75; .5')
alpha = 10
a = np.linalg.inv(np.dot(X.T, X) + alpha*np.identity(3))
b = np.dot(X.T, Y)
w = np.dot(a, b)
sqrNormW = np.linalg.norm(w)**2
print(w)
print(sqrNormW)
# smaller sqaured norm since a larger alpha value puts more emphasis on minimizing
# the squared norm.
```

```python
def f(net):
    return 1 / (1 + np.exp(-net))

x = np.matrix('.9,.8; .5,.6; .7,.7; .6,.5')
y = np.matrix('.95; .5; .75; .5')
z = np.matrix('0,0')
o = np.matrix('0')
bh = np.matrix('0,0')
bo = np.matrix('0')
wh = np.matrix('0,0;0,0')
wo = np.matrix('0,0')

for i in range(2):
    for i in range(4):
        z = f(bh + (wh*x[i,:].T).T)
        o = f(bo + (wo*z.T).T)

        do = o*(1-o)*(o-y[i,:])
        dh = (z.T*(1-z)*(wo.T*do)).T

        bo = bo - do
        bh = bh - dh

        wo = wo - (z.T*do).T
        wh = wh - (x[i,:].T*dh).T

        print('\n#####################################\n')
        print('do')
        print(do)
        print('\ndh')
        print(dh)
        print('\nbo')
        print(bo)
        print('\nbh')
        print(bh)
        print('\nwo')
        print(wo)
        print('\nwh')
        print(wh)
        print('\nz')
        print(z)
        print('\no')
        print(o)
```

```python
def f(net):
    return 1 / (1 + np.exp(-net))

x = np.matrix('.9,.8; .5,.6; .7,.7; .6,.5').T
y = np.matrix('.95; .5; .75; .5').T
z = np.matrix('0,0').T
o = 0
bh = np.matrix('0,0').T
bo = 0
wh = np.matrix('0,0;0,0')
wo = np.matrix('0,0').T

for i in range(2):
    for i in range(4):
        z = f(bh + np.dot(wh.T, x[:,i]))
        o = f(bo + np.dot(wo.T, z))

        do = o*(1-o)*(o-y[:,i])
        dh = np.dot(z.T, (1-z))[0,0]*(wo*do)

        bo = bo - do
        bh = bh - dh

        wo = wo - z*do
        wh = wh - np.dot(x[:,i], dh.T)

        print('\n#####################################\n')
        print('do')
        print(do)
        print('\ndh')
        print(dh)
        print('\nbo')
        print(bo)
        print('\nbh')
        print(bh)
        print('\nwo')
        print(wo)
        print('\nwh')
        print(wh)
        print('\nz')
        print(z)
        print('\no')
        print(o)
```
